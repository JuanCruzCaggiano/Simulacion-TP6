```python
import pandas as pd
```

```python
df = pd.read_csv("customer_support_tickets.csv")
df.drop(columns=["Ticket ID","Customer Name","Customer Email","Customer Age","Customer Gender","Product Purchased","Date of Purchase","Ticket Type","Ticket Subject","Ticket Description","Resolution","Ticket Channel"], inplace=True)
#filtramos los tickets que est√©n cerrado, con tiempo de inicio y cierre.
df_filtered = df[df['Ticket Status'] == 'Closed']
df_filtered = df_filtered[df_filtered['Time to Resolution'].notna()]
df_filtered = df_filtered[df_filtered['First Response Time'].notna()]
#una vez filtrado el campo ticket status no nos interesa mas
df_filtered = df_filtered.drop(columns=["Ticket Status"])
```
